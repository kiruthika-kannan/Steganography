```python
import numpy as np
from matplotlib import pyplot as plt
import cv2

from least_significant_bit_embedding import embedding,preprocessing
from least_significant_bit_extraction import extraction
```

```python
noOfReplaceBits = 3
coverImage = preprocessing(cv2.imread('./images/img5.jpg'))
payloadImage = preprocessing(cv2.imread('./images/img3.png'))
stegoImage = embedding(coverImage,payloadImage,noOfReplaceBits,True)
cv2.imwrite('./images/stegoImage.png',stegoImage)
```

```python
stegoImage = preprocessing(cv2.imread('./images/stegoImage.png'))
coverImageExtracted,payloadImageExtracted = extraction(stegoImage,noOfReplaceBits,True)
MSECover = np.mean(np.square(coverImage - coverImageExtracted))
MSEPayload = np.mean(np.square(payloadImage - payloadImageExtracted))
plt.show()
print("Mean Square Error in Cover Image Recovery: ", MSECover)
print("Mean Square Error in Payload Image Recovery: ", MSEPayload)
```

```python
for i in range(1,8):
    print('___________________________________________________________________________')
    print("No of replaced bits: ",i)
    noOfReplaceBits = i
    stegoImage = embedding(coverImage,payloadImage,noOfReplaceBits,True)
    cv2.imwrite('./images/stegoImage'+str(i)+'.png',stegoImage)
    stegoImage = preprocessing(cv2.imread('./images/stegoImage'+str(i)+'.png'))
    coverImageExtracted,payloadImageExtracted = extraction(stegoImage,noOfReplaceBits,True)
    print("Mean Square Error in Cover Image Recovery: ", MSECover)
    print("Mean Square Error in Payload Image Recovery: ", MSEPayload)
    
```
